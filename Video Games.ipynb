## Video Games

```python
import pandas as pd 
```

```python
djs = pd.read_csv('dados/vgsales.csv')
djs.head(5)
```

## Idade dos Jogos 😎

```python
from datetime import datetime
ano_atual = datetime.today().year
djs['Idade_do_jogo'] = ano_atual - djs.Year
djs
```

## Trocando nomes das colunas 

```python
dataset = djs.rename(columns = {'Year': 'Ano', 'Global_Sales': 'Global Vendas', 'Genre': 'Genero', 'Name': 'Nome', 'Publisher': 'Editora', 'Platform': 'Plataforma'})
```

```python
dataset
```

### Valor Maximo Da Editora Activision No PC 

```python
activision = (dataset['Editora'] == 'Activision') & (dataset['Genero'] == 'Shooter') & (dataset['Plataforma'] == 'PC')
Editora_Activision = dataset[activision]
Editora_Activision
```

```python
Edit =  Editora_Activision.groupby('Nome')
Edit
```

```python
Edit.describe()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20,10))
```

```python
fig = Edit['Global Vendas'].max().plot.bar(color =  'black')
fig.set_ylabel('Valor Maximo')
fig.set_title('Valor Medio', {'fontsize': 22})
```

### Selecionando os Jogos FIFA, GTA V e Call Of Duty 

```python
selecao = (dataset['Nome'] == 'FIFA 14') | (dataset['Nome'] == 'FIFA 15') |  (dataset['Nome'] == 'FIFA 16') | (dataset['Nome'] == 'Grand Theft Auto V') | (dataset['Nome'] == 'Call of Duty: Black Ops 3') & (dataset['Ano'] == 2015)
plat = dataset[selecao]
plat
```

## Agrupando os nomes

```python
game = plat.groupby('Nome')
game
```

##  Estatistica dos Jogos

```python
game[['Global Vendas']].describe()
```

## Importando o matplotlib
