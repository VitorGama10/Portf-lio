# Funçao DEF 😀

##### Fazer uma divisão dos kms dos carros ate 2022

```python
def km_media(dataset, ano_atual):
  for item in dataset.items():
    result = item[1]['km'] / (ano_atual - item[1]['ano'])
    print(result)
```

```python
km_media(kms, 2022)
```

### Adicionando no Dict

```python
def km_media(dataset, ano_atual):
    result = {}
    for item in dataset.items():
        media = item[1]['km'] / (ano_atual - item[1]['ano'])
        item[1].update({'km_media': media})
        result.update({item[0]: item[1]}) # Para Adicionar no dict
      
      
    return result
```

```python
km_media(kms, 2022)
```

# Criar DataFrame

```python
k1 = pd.DataFrame(km_media(kms, 2022)).T.round(2)
k1
```

##Total de Motores 👍

```python
motor = grupo_carros['Motor'].value_counts().to_frame().reset_index()
motores = motor.rename(columns = {'index': 'Motores', 'Motor': 'Quantidades'})
motores
```
